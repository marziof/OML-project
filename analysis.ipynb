```python
import torch
import pandas as pd
import copy
from model import*

def test(model, loader):
    model.eval()
    correct = 0
    total = 0

    with torch.no_grad():
        for x, y in loader:
            out = model(x)
            preds = out.argmax(dim=1)
            correct += (preds == y).sum().item()
            total += y.size(0)

    return correct / total


criterion = torch.nn.CrossEntropyLoss()
results_MNIST_df = pd.DataFrame(columns=["Optimizer", "Train Loss", "Test Loss", "learning_rate", "Test Accuracy", "train_loss_curve", "test_loss_curve"])

lrs = [0.01, 0.05, 0.1]
n_epochs = 10
# track loss (test and train) for each epoch, for each optimizer, for each lr

for lr in lrs:
    print(f"Running with learning rate: {lr}")
    # --- same initialization (IMPORTANT) ---
    base_model = SimpleMLP()
    model_lh = copy.deepcopy(base_model)
    model_adam = copy.deepcopy(base_model)

    opt_lh = SGD_LH(model_lh.parameters(), lr=lr)
    opt_adam = torch.optim.Adam(model_adam.parameters(), lr=lr)

    train_loss_curve_lh = []
    train_loss_curve_adam = []
    test_loss_curve_lh = []
    test_loss_curve_adam = []


    # --- training ---
    for epoch in range(n_epochs):
        if (epoch + 1) % 5 == 0:
            print(f"Epoch {epoch+1}/{n_epochs}")
        train_loss_lh = train(model_lh, opt_lh, train_loader, criterion)
        train_loss_adam = train(model_adam, opt_adam, train_loader, criterion)

        train_loss_curve_lh.append(train_loss_lh)
        train_loss_curve_adam.append(train_loss_adam)

        # --- evaluation ---
        test_loss_lh = evaluate(model_lh, test_loader, criterion)
        test_loss_adam = evaluate(model_adam, test_loader, criterion)
        test_loss_curve_lh.append(test_loss_lh)
        test_loss_curve_adam.append(test_loss_adam)

    accuracy_lh = test(model_lh, test_loader)
    accuracy_adam = test(model_adam, test_loader)

    # --- logging ---
    results_MNIST_df.loc[len(results_MNIST_df)] = [ "SGD_LH", train_loss_lh, test_loss_lh, lr, accuracy_lh, train_loss_curve_lh, test_loss_curve_lh]

    results_MNIST_df.loc[len(results_MNIST_df)] = [ "Adam", train_loss_adam, test_loss_adam, lr, accuracy_adam, train_loss_curve_adam, test_loss_curve_adam]
```

```python
import torch
import math
from torch import nn
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as T
from torchvision.datasets import CIFAR10
import torchvision.models as models
import torch.nn.functional as F

class SimCLRTransform:

    def __init__(self, size=32, s=0.5, blur_p=0.5):
        color_jitter = T.ColorJitter(0.8*s, 0.8*s, 0.8*s, 0.2*s)
        k = 3 if size <= 32 else 5
        base = [
            T.RandomResizedCrop(size=size, scale=(0.2, 1.0)),
            T.RandomHorizontalFlip(),
            T.RandomApply([color_jitter], p=0.8),
            T.RandomGrayscale(p=0.2),
            #T.RandomApply([T.GaussianBlur(kernel_size=k, sigma=(0.1, 2.0))], p=blur_p),
            T.ToTensor()
        ]
        self.train_transform = T.Compose(base)

    def __call__(self, x):
        return self.train_transform(x), self.train_transform(x)


def simclr_collate_fn(batch):
    xs1, xs2, ys = [], [], []
    for (x1, x2), y in batch:
        xs1.append(x1)
        xs2.append(x2)
        ys.append(y)
    return torch.stack(xs1), torch.stack(xs2), torch.tensor(ys)

simclr_transform = SimCLRTransform(size=32)

default_transform = T.Compose([
    T.ToTensor(),
])

train = CIFAR10('./data', train=True, download=True, transform=default_transform)
test = CIFAR10('./data', train=False, download=True, transform=default_transform)

train_ds = CIFAR10(root="./data", train=True, download=True, transform=simclr_transform)
test_ds  = CIFAR10(root="./data", train=False, download=True, transform=simclr_transform)

train_loader = DataLoader(train_ds, batch_size=32, num_workers=0, pin_memory=True, collate_fn=simclr_collate_fn)
test_loader  = DataLoader(test_ds,  batch_size=32, num_workers=0, pin_memory=True, collate_fn=simclr_collate_fn)

train_ds_noaugment = CIFAR10(root="./data", train=True, download=True, transform=default_transform)
test_ds_noaugment  = CIFAR10(root="./data", train=False, download=True, transform=default_transform)

train_loader_noaugment = DataLoader(train_ds_noaugment, batch_size=32, shuffle=False, num_workers=0, pin_memory=True)
test_loader_noaugment  = DataLoader(test_ds_noaugment,  batch_size=32, shuffle=False, num_workers=0, pin_memory=True)
len(train_ds), len(test_ds)
```

```python
class SimCLR(nn.Module):

    def __init__(self, proj_dim=128, hidden=2048):
        super().__init__()
        enc = models.resnet18(weights=None)
        enc.conv1 = nn.Conv2d(3, 64, 3, 1, 1, bias=False)
        enc.maxpool = nn.Identity()
        enc.fc = nn.Identity()
        self.encoder = enc

        self.projector = nn.Sequential(
            nn.Linear(512, hidden),
            nn.ReLU(inplace=True),
            nn.Linear(hidden, proj_dim) # (bs, proj_dim)
        )

    def normalize(self, x, eps=1e-8):
        return x / (x.norm(dim=-1, keepdim=True) + eps)

    def forward(self, x):
        h = self.encoder(x)
        z = self.normalize(self.projector(h))
        return h, z


#@torch.compile
def nt_xent(z1, z2, tau=0.5):
    B, d = z1.shape
    z = torch.cat([z1, z2], dim=0)              # (2B, d)
    sim = (z @ z.t()) / tau                     # (2B, 2B)
    mask = torch.eye(2*B, dtype=torch.bool, device=z.device)
    sim.masked_fill_(mask, -1e9)
    targets = torch.arange(B, device=z.device)
    targets = torch.cat([targets + B, targets], dim=0)
    return F.cross_entropy(sim, targets)

from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics as skm

def run_knn_probe(train_features, train_labels, test_features, test_labels):
    """
    Runs a k-NN probe on the given features and labels.
    train_features: (num_train_samples, feature_dim)
    train_labels: (num_train_samples,)
    test_features: (num_test_samples, feature_dim)
    test_labels: (num_test_samples,)
    returns: accuracy (float)
    """
    knn = KNeighborsClassifier(n_neighbors=5, n_jobs=-1)
    knn.fit(train_features, train_labels)
    test_preds = knn.predict(test_features)
    accuracy = skm.accuracy_score(test_labels, test_preds)
    return accuracy

#device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")
print(device)


def extract_features_and_labels(model, dataloader, normalize=False):
    """
    Extract features and labels from a dataloader using the given model.
    model: an encoder model taking as input a batch of images (batch_size, channels, height, width) and outputing either a batch of feature vectors (batch_size, feature_dim) or a list/tuple in which the first element is the batch of feature vectors (batch_size, feature_dim)
    dataloader: a PyTorch dataloader providing batches of (images, labels)
    returns: features (num_samples, feature_dim), labels (num_samples,)
    """
    features = []
    labels = []

    for batch in dataloader:
        x, y = batch
        x = x.to(device)
        with torch.no_grad():
            feat = model(x)
            if type(feat) is tuple or type(feat) is list:
                repr = feat[0]
                feat = repr
        features.append(feat.cpu())
        labels.append(y)

    features = torch.cat(features, dim=0)
    labels = torch.cat(labels, dim=0)

    if normalize:
        features = F.normalize(features, dim=1)

    return features, labels
```

```python
import copy
from tqdm import tqdm

total_epochs = 5
warmup_epochs = 5
lr=0.1

def lr_lambda(epoch):
    if epoch < warmup_epochs:
        return (epoch + 1) / float(warmup_epochs)
    t = (epoch - warmup_epochs) / float(total_epochs - warmup_epochs)
    return 0.0 + 0.5 * (1 - 0.0) * (1 + math.cos(math.pi * t))

#optimizer = torch.optim.AdamW(simclr_model.parameters(), lr=0.6, weight_decay=0.0)
#scheduler = torch.optim.lr_scheduler.LambdaLR(optimizer, lr_lambda)

base_model = SimCLR(proj_dim=128).to(device)

model_lh = copy.deepcopy(base_model)
model_adam = copy.deepcopy(base_model)

# compile AFTER copying
#model_lh = torch.compile(model_lh)
#model_adam = torch.compile(model_adam)

opt_lh = SGD_LH(model_lh.parameters(), lr=lr)
opt_adam = torch.optim.Adam(model_adam.parameters(), lr=lr)

all_accuracies_adam = []
all_accuracies_lh = []
train_loss_curve_lh = []
train_loss_curve_adam = []

for epoch in tqdm(range(total_epochs)):
    #print(f"Epoch {epoch+1}/{total_epochs}")
    model_adam.train()
    model_lh.train()
    avg_loss_adam = 0.
    avg_loss_lh = 0.
    for i, (x1, x2, y) in enumerate(tqdm(train_loader, leave=False)):
        if i >= len(train_loader) // 2:
            break
        #print("In training loop...")
        x1, x2 = x1.to(device), x2.to(device)
        x = torch.cat([x1, x2], dim=0)
        
        _, proj_adam = model_adam(x)
        proj1_adam, proj2_adam = proj_adam.chunk(2)

        # _, proj1_adam = model_adam(x1)
        # _, proj2_adam = model_adam(x2)
        
        loss_adam = nt_xent(proj1_adam, proj2_adam, tau=0.5)
        #print(f"Adam loss: {loss_adam.item():.4f}")

        _, proj_lh = model_lh(x)
        proj1_lh, proj2_lh = proj_lh.chunk(2)
        # _, proj1_lh = model_lh(x1)
        # _, proj2_lh = model_lh(x2)
        loss_lh = nt_xent(proj1_lh, proj2_lh, tau=0.5)
        #print(f" LH loss: {loss_lh.item():.4f}")
        if i % 50 == 0:
            tqdm.write(f"At batch {i}/{len(train_loader)//2}: Adam: {loss_adam.item():.4f} | LH: {loss_lh.item():.4f}")

        opt_adam.zero_grad()
        loss_adam.backward()
        opt_adam.step()
        #print("Adam step done.")
        opt_lh.zero_grad()
        loss_lh.backward()
        opt_lh.step()
        #print("LH step done.")

        avg_loss_adam += loss_adam.item()
        avg_loss_lh += loss_lh.item()

    #scheduler.step()
    model_lh.eval()
    model_adam.eval()
    train_features_lh, train_labels_lh = extract_features_and_labels(model_lh, train_loader_noaugment, normalize=True)
    test_features_lh, test_labels_lh = extract_features_and_labels(model_lh, test_loader_noaugment, normalize=True)
    train_features_adam, train_labels_adam = extract_features_and_labels(model_adam, train_loader_noaugment, normalize=True)
    test_features_adam, test_labels_adam = extract_features_and_labels(model_adam, test_loader_noaugment, normalize=True)
    acc_adam = run_knn_probe(train_features_adam, train_labels_adam, test_features_adam, test_labels_adam)
    acc_lh = run_knn_probe(train_features_lh, train_labels_lh, test_features_lh, test_labels_lh)
    avg_loss_adam = avg_loss_adam / len(train_loader)
    avg_loss_lh = avg_loss_lh / len(train_loader)
    print(f"Epoch {epoch+1}, Average Loss: {avg_loss_adam:.4f}, kNN Accuracy: {acc_adam:.4f}")
    all_accuracies_adam.append(acc_adam)
    all_accuracies_lh.append(acc_lh)
    train_loss_curve_lh.append(avg_loss_lh)
    train_loss_curve_adam.append(avg_loss_adam)

results_CIFAR_df = pd.DataFrame(columns=["Optimizer", "Train Loss", "Test Loss", "learning_rate", "Test Accuracy", "train_loss_curve", "test_loss_curve"])
results_CIFAR_df.loc[len(results_CIFAR_df)] = [ "SGD_LH", train_loss_lh, test_loss_lh, lr, accuracy_lh, train_loss_curve_lh, test_loss_curve_lh]
results_CIFAR_df.loc[len(results_CIFAR_df)] = [ "Adam", train_loss_adam, test_loss_adam, lr, accuracy_adam, train_loss_curve_adam, test_loss_curve_adam]
```
